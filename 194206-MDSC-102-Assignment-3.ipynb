# Probability Distributions

# 1.Binomial Distribution 


```python
import numpy as np
import matplotlib.pyplot as plt

def binomial_pmf(p, r, n):
    return (p ** r) * ((1 - p) ** (n - r))

p = 0.2
n = 30

# Calculate the probabilities for all possible values of r from 0 to n
r_values = np.arange(n + 1)
probabilities = binomial_pmf(p, r_values, n)

# Create a bar plot to visualize the distribution
plt.bar(r_values, probabilities, color='skyblue', edgecolor='black', linewidth=1.2)
plt.xlabel('Number of Successes (r)')
plt.ylabel('Probability')
plt.title('Binomial Distribution')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define a lambda function for binomial PMF
binomial_pmf = lambda p, r_values, n: (p ** r_values) * ((1 - p) ** (n - r_values))

p = 0.2
n = 30
r_values = np.arange(n + 1)

# Calculate the probabilities for all possible values of r from 0 to n
probabilities = binomial_pmf(p, r_values, n)

# Create a bar plot to visualize the distribution
plt.bar(r_values, probabilities, color='skyblue', edgecolor='black', linewidth=1.2)
plt.xlabel('Number of Successes (r)')
plt.ylabel('Probability')
plt.title('Binomial Distribution ')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

# 2.Poisson distribution

```python
def poisson_pmf(k, lambda_p):
    return (np.exp(-lambda_p) * (lambda_p ** k)) / np.math.factorial(k)

def plot_PD(lambda_p, max_value):
    # Generate the Poisson distribution
    x = np.arange(0, max_value + 1)
    poisson_dist = np.array([poisson_pmf(k, lambda_p) for k in x])
    plt.bar(x, poisson_dist, color='skyblue', edgecolor='black', linewidth=1.2)
    plt.grid(axis='y', linestyle='--', alpha=1)
    plt.xlabel("Number of Events")
    plt.ylabel("Probability")
    plt.title(f"Poisson Distribution (lambda={lambda_param})")
    plt.show()
#lambda_p = 0.01
lambda_p = 5
#lambda_p = 1
max_value_to_plot = 10 
plot_PD(lambda_p, max_value_to_plot)
```

# 3.Exponential f(x) = λ * exp(-λ * x)

```python
lambda_p = 0.5
x = np.linspace(0, 10, 10)
exponential_dist = lambda_p * np.exp(-lambda_p * x) #f(x) = λ * exp(-λ * x)

plt.bar(x, exponential_dist, color='skyblue', linewidth=1.2)
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title(f"Exponential Distribution (λ is {lambda_p})")
plt.show()
```

# 4.Normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt

def normal_distribution(x, m, s):
    return (1 / (s * np.sqrt(2 * np.pi))) * np.exp(-0.5 * ((x - m) / s) ** 2)

m = 3  # mean
s = 2  # standard_dev

x = np.linspace(-5, 5, 1000)
pdf_values = normal_distribution(x, m, s)

plt.plot(x, pdf_values, color='skyblue', linewidth=2)
plt.xlabel("x")
plt.ylabel("Probability Density")
plt.title(f"Normal Distribution (mean={m}, std={s})")
plt.grid(True)
plt.show()
```

# 5.Gamma Distribution
